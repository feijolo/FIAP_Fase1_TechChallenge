```python
# 5) Preparar dataset — exemplo genérico usando uid_label_map.csv (uid,label,patient_id) e imagens em data/cbis_ddsm_demo/
import pandas as pd
from sklearn.model_selection import GroupShuffleSplit
UID_CSV = os.path.join(PROJECT_DIR, 'data', 'uid_label_map.csv')
df = pd.read_csv(UID_CSV)
df.head()
```

```python
# 6) Exemplo: separar por paciente (70% treino, 15% val, 15% test)
splitter = GroupShuffleSplit(n_splits=1, train_size=0.7, random_state=42)
train_idx, rest_idx = next(splitter.split(df, groups=df['patient_id']))
rest = df.iloc[rest_idx].reset_index(drop=True)
splitter2 = GroupShuffleSplit(n_splits=1, train_size=0.5, random_state=42)
val_idx, test_idx = next(splitter2.split(rest, groups=rest['patient_id']))
train_df = df.iloc[train_idx].reset_index(drop=True)
val_df = rest.iloc[val_idx].reset_index(drop=True)
test_df = rest.iloc[test_idx].reset_index(drop=True)
print(len(train_df), len(val_df), len(test_df))
```

```python
# 7) Dataset e DataLoader (PyTorch) — adapta conforme sua estrutura de nomes/folders
import torch
from torch.utils.data import Dataset, DataLoader
from PIL import Image
from torchvision import transforms

IMG_DIR = os.path.join(PROJECT_DIR, 'data', 'cbis_ddsm_demo')
IMG_SIZE = 224
transform_train = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])
transform_val = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

class SimpleImageDataset(Dataset):
    def __init__(self, df, img_dir, transform=None):
        self.df = df
        self.img_dir = img_dir
        self.transform = transform
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = os.path.join(self.img_dir, row['uid'])  # ajuste se uid inclui extensão
        if not os.path.exists(img_path):
            # tentar adicionar extensão .png/jpg
            if os.path.exists(img_path + '.png'): img_path += '.png'
            elif os.path.exists(img_path + '.jpg'): img_path += '.jpg'
        img = Image.open(img_path).convert('RGB')
        if self.transform: img = self.transform(img)
        label = int(row['label'])
        return img, label

train_ds = SimpleImageDataset(train_df, IMG_DIR, transform_train)
val_ds = SimpleImageDataset(val_df, IMG_DIR, transform_val)
test_ds = SimpleImageDataset(test_df, IMG_DIR, transform_val)

train_loader = DataLoader(train_ds, batch_size=32, shuffle=True, num_workers=2)
val_loader = DataLoader(val_ds, batch_size=32, shuffle=False, num_workers=2)
test_loader = DataLoader(test_ds, batch_size=32, shuffle=False, num_workers=2)
```

```python
# 8) Criar modelo (transfer learning com timm)
import timm
from torch import nn, optim
device = 'cuda' if torch.cuda.is_available() else 'cpu'
num_classes = len(df['label'].unique())
model = timm.create_model('resnet18', pretrained=True, num_classes=num_classes)
model = model.to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
```

```python
# 9) Treino simplificado — salve checkpoints em PROJECT_DIR + '/checkpoints'
from tqdm import tqdm
os.makedirs(os.path.join(PROJECT_DIR,'checkpoints'), exist_ok=True)
n_epochs = 5  # ajuste conforme recursos
best_val = 1e9
for epoch in range(n_epochs):
    model.train()
    running_loss = 0.0
    for imgs, labels in tqdm(train_loader):
        imgs = imgs.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * imgs.size(0)
    train_loss = running_loss / len(train_loader.dataset)
    # validação
    model.eval()
    val_loss = 0.0
    correct = 0
    with torch.no_grad():
        for imgs, labels in val_loader:
            imgs = imgs.to(device)
            labels = labels.to(device)
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * imgs.size(0)
            preds = outputs.argmax(dim=1)
            correct += (preds == labels).sum().item()
    val_loss = val_loss / len(val_loader.dataset)
    val_acc = correct / len(val_loader.dataset)
    print(f'Epoch {epoch+1}/{n_epochs} — train_loss: {train_loss:.4f} val_loss: {val_loss:.4f} val_acc: {val_acc:.4f}')
    # salvar checkpoint
    ckpt_path = os.path.join(PROJECT_DIR, 'checkpoints', f'model_epoch_{epoch+1}.pt')
    torch.save({'epoch': epoch+1, 'model_state': model.state_dict(), 'optimizer_state': optimizer.state_dict()}, ckpt_path)
```

```python
# 10) Avaliação final: gerar relatório, matriz de confusão e ROC
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

model.eval()
all_probs = []
all_preds = []
all_labels = []
with torch.no_grad():
    for imgs, labels in test_loader:
        imgs = imgs.to(device)
        outputs = model(imgs)
        probs = torch.softmax(outputs, dim=1).cpu().numpy()
        preds = outputs.argmax(dim=1).cpu().numpy()
        all_probs.append(probs)
        all_preds.append(preds)
        all_labels.append(labels.numpy())
all_probs = np.vstack(all_probs)
all_preds = np.hstack(all_preds)
all_labels = np.hstack(all_labels)
# salvar relatório
report = classification_report(all_labels, all_preds, output_dict=True)
import json
with open(os.path.join(PROJECT_DIR,'reports','image_colab_classification_report.json'),'w') as f:
    json.dump(report, f, indent=2)
# confusion matrix
cm = confusion_matrix(all_labels, all_preds)
os.makedirs(os.path.join(PROJECT_DIR,'reports'), exist_ok=True)
plt.figure(figsize=(6,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predito')
plt.ylabel('Verdadeiro')
plt.savefig(os.path.join(PROJECT_DIR,'reports','colab_confusion_matrix.png'))
plt.close()
# ROC (one-vs-rest se multi-classe) — exemplo binário
if all_probs.shape[1] == 2:
    fpr, tpr, _ = roc_curve(all_labels, all_probs[:,1])
    roc_auc = auc(fpr,tpr)
    plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.3f}')
    plt.plot([0,1],[0,1],'--',color='gray')
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.legend()
    plt.savefig(os.path.join(PROJECT_DIR,'reports','colab_roc.png'))
    plt.close()
```

```python
# 11) Grad-CAM (exemplo) — salva imagens no diretório de reports
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.image import show_cam_on_image
import numpy as np

target_layer = model.layer4 if hasattr(model,'layer4') else model.get_classifier()
cam = GradCAM(model=model, target_layers=[target_layer], use_cuda=(device=='cuda'))
os.makedirs(os.path.join(PROJECT_DIR,'reports','gradcam'), exist_ok=True)
# pegar alguns exemplos do test_ds
for i in range(min(8, len(test_ds))):
    img, label = test_ds[i]
    input_tensor = img.unsqueeze(0).to(device)
    grayscale_cam = cam(input_tensor=input_tensor, targets=None)[0]
    # preparar imagem RGB para sobreposição (norm back to 0-1)
    img_np = img.permute(1,2,0).numpy()
    img_np = (img_np - img_np.min()) / (img_np.max()-img_np.min()+1e-8)
    visualization = show_cam_on_image(img_np, grayscale_cam, use_rgb=True)
    from PIL import Image
    Image.fromarray(visualization).save(os.path.join(PROJECT_DIR,'reports','gradcam',f'gradcam_{i}.png'))
```
